```python
#plot the CPU and GPU running time with different N size

import matplotlib.pyplot as plt
import numpy as np

def plot_data(data):
    plt.plot(data[:,0], data[:,1], label='CPU')
    plt.plot(data[:,0], data[:,2], label='GPU')
    plt.xlabel('N')
    plt.ylabel('Time (ms)')
    plt.title('CPU vs GPU running time')
    plt.legend()
```

```python
#N = { 1-500 }
data_1_500 = np.loadtxt('logger_1_500.csv', delimiter=',', skiprows=1)
data_500_1000 = np.loadtxt('logger_500_1000.csv', delimiter=',', skiprows=1)
data_1_1000 = np.concatenate((data_1_500, data_500_1000), axis=0)

plot_data(data_1_1000)
```

```python
#N = { 500-1000 }
data_500_1000 = np.loadtxt('logger_500_1000.csv', delimiter=',', skiprows=1)
plot_data(data_500_1000)
```

```python
#N = { 500-1000 }
data_1000 = np.loadtxt('logger_1000_10000000.csv', delimiter=',', skiprows=1)
plot_data(data_1000)
```
